```python
import os, sys, subprocess, glob, numpy as np, pickle as pkl
import matplotlib.pyplot as plt, itertools

from scipy.optimize import curve_fit

%matplotlib inline
```

```python
# load responses
sim_slow_dict = pkl.load(open('implicit_slow_response_fluct_correction_cut-1.3.pkl', 'rb'))
sim_fast_dict = pkl.load(open('implicit_fast_cut-1.3_response.pkl', 'rb'))

# make slow dict from experimental data
exp_slow_dict = {}
for line in open('experimental_slow_responses_kd.txt', 'r'):  # plot kd or kobs
    
    line = line.replace('\n', '')
    if line in sim_slow_dict:
        print(line)
        seq = line
        temp_list = []
        exp_list = []
    
    elif line == '':

        exp_slow_dict[seq] = (temp_list, exp_list) 
        
    else:
        [temp, exp] = line.split()
        temp_list.append(int(temp)+273)
        exp_list.append(float(exp))

        
# make fast dict from experimental data
exp_fast_dict = {}
for line in open('experimental_fast_responses.txt', 'r'):
    
    line = line.replace('\n', '')
    if line in sim_fast_dict:
        print(line)
        seq = line
        temp_list = []
        exp_list = []
    
    elif line == '':
        exp_fast_dict[seq] = (temp_list, exp_list) 
    
    else:
        [temp, exp] = line.split()
        temp_list.append(int(temp)+273)
        exp_list.append(float(exp))
```

```python
# plot fast response wil temp shift and acceleration factor

plt.rcParams.update({'font.size': 20})
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

color_dict = {'AT-all':'orange', 'GC-end':'red', 'GC-core':'blue', 'GC-mix':'green'}

def get_axis_limits(ax, x, y):
    return ax.get_xlim()[1]*x, ax.get_ylim()[1]*y
    
# specify shift and acceleration for fast and slow responses
t_shift_slow = 4      #-4    #k
t_acc_slow   = 10 #10.0  #acceleration factor

t_shift_fast = 4   #k
t_acc_fast   = 120  #acceleration factor

axes = ax[1]
for seq, color in color_dict.items():
    
    # load simulation fits
    temp_list, exp_list, sig_list = sim_fast_dict[seq]
    exps = np.array(exp_list[1:]) / t_acc_fast
    temps = np.array(temp_list[1:]) + t_shift_fast
    
    axes.scatter(temps, exps, c=color)
    axes.plot(temps, exps, label=seq+' sim', c=color)
    
    # load experimental data
    temps, exps = exp_fast_dict[seq]
    temps = np.array(temps) 
    exps = np.array(exps)
    
    axes.scatter(temps, exps, c=color)
    axes.plot(temps, exps, label=seq+' exp', linestyle='dashed', c=color)
    
axes.set_ylabel('$k_d^\mathrm{fast}$     ($\mathrm{s}^{-1}$)')
axes.set_xlabel('Temperature (K)')
axes.ticklabel_format(axis='y', style='sci')
axes.set_title(f'Fast Responses') #\n({int(t_acc_fast)}x accelerated)')
axes.legend()

axes = ax[0]
for seq, color in color_dict.items():
    
    # load simulation fits
    temp_list, exp_list, sig_list = sim_slow_dict[seq]
    exps = np.array(exp_list)[1:]  / t_acc_slow  #acceleration factor
    temps = np.array(temp_list)[1:] + t_shift_slow  # temp shift
    
    axes.scatter(temps, exps, c=color)
    axes.plot(temps, exps, label=seq+' sim', c=color)
    
    # load simulation data
    temps, exps = exp_slow_dict[seq]
    temps = np.array(temps)
    exps = np.array(exps)
    
    axes.scatter(temps, exps, c=color)
    axes.plot(temps, exps, label=seq+' exp', linestyle='dashed', c=color)
    
#axes.annotate('b', xy=get_axis_limits(axes, 0.8, 0.8))
axes.set_ylabel('$k_d^\mathrm{slow}$     ($\mathrm{s}^{-1}$)')
axes.set_xlabel('Temperature (K)')
axes.set_title(f'Slow Responses') #\n({int(t_acc_slow)}x accelerated)')
axes.ticklabel_format(axis='y', style='sci', scilimits=(10, 12))
axes.legend()

# make log to show exps
axes.set_yscale('log')

plt.subplots_adjust(hspace=0.4)

save_path = '/home/mikejones/scratch-midway2/srv/implicit/paper_figs/'
plt.savefig(f'{save_path}_fast_{t_shift_fast}_slow_{t_shift_slow}_responses.pdf', dpi=600, bbox_inches='tight')
plt.savefig(f'{save_path}_fast_{t_shift_fast}_slow_{t_shift_slow}_responses.png', dpi=600, bbox_inches='tight')
```
